```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [25,75]
plt.pie(data)
plt.show()
```

<b>Important: Notice the use of plt.pie() to plot the pie chart, NOT plt.plot()</b><br/>
<b>Note: It is important to set axis as equal so that our pie chart will shape as circle!</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [25,75]
plt.pie(data)
plt.axis('equal') 
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [100, 50, 150]
plt.pie(data)
plt.axis('equal')
plt.show()
```

Below are the useful parameters for styling basic pie charts:
* labels display the label of each data point
* title add the title of the chart
* colors assigned the specified color to each data point
* explode to split one slice from the pie
* to have shawdow, set shadow equals True
* autopct

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [120, 50, 150]
lbl = ['Coffe', 'Tea', 'Milk']
plt.pie(data, labels=lbl)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [120, 50, 150]
lbl = ['Coffe', 'Tea', 'Milk']
col = ['brown', 'orange', 'gray']
plt.pie(data, labels=lbl, colors=col)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [120, 50, 150]
lbl = ['Coffe', 'Tea', 'Milk']
col = ['brown', 'orange', 'gray']
spl = [0,0.1,0]
plt.pie(data, labels=lbl, colors=col, explode=spl)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

<b>Note: to use explode, we need to feed the data into array-like format</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [120, 50, 150]
lbl = ['Coffe', 'Tea', 'Milk']
col = ['brown', 'orange', 'gray']
spl = [0,0.2,0]
plt.pie(data, labels=lbl, colors=col, explode=spl, shadow=True)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data = [120, 50, 150]
lbl = ['Coffe', 'Tea', 'Milk']
col = ['brown', 'orange', 'gray']
spl = [0,0.2,0]
plt.pie(data, labels=lbl, colors=col, explode=spl, shadow=True, autopct='%1.2f%%')
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```
