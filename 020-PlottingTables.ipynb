Before plotting the table, there are few parameters we should be familiar with:
* figure - create a figure to plot
* subplots - add an axes to the figure, value 1,1 means 1x1 grid
* axis('off') to turn off the x, y axis values (left & bottom)
* set_aspect accepts three different param, i.e. auto, equal and numeric value (cell size)
* scale  - scale column widths and row heights

```python
%matplotlib inline
import matplotlib.pyplot as plt

data=[
    [1, 2, 3, 4, 5],
    [2, 4, 6, 8, 10],
    [1, 3, 5, 7, 9]
]

#fig = plt.figure()
#ax = fig.subplots(1,1)
fig, ax = plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)

table = ax.table(cellText=data)
table.scale(1, 10)
```

<b>Note: To plot table, we use ax.table by passing our data into cellText parameter</b>

Below are few additional parameters for table:
* colLabels - list of string, the text of the column header cells
* cellLoc - the alignment of the text within the cells
* loc - the position of the cell with respect to ax
* rowLabels - list of string, the text of the row header cells

```python
%matplotlib inline
import matplotlib.pyplot as plt

data=[
    [1, 2, 3, 4, 5],
    [2, 4, 6, 8, 10],
    [1, 3, 5, 7, 9]
]
cols = ['ColOne', 'ColTwo', 'ColThree', 'ColFour', 'ColFive']

fig = plt.figure()
ax = fig.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)

table = ax.table(cellText=data, colLabels=cols, loc='upper center',cellLoc='center')
table.scale(1, 10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data=[
    [1, 2, 3, 4, 5],
    [2, 4, 6, 8, 10],
    [1, 3, 5, 7, 9]
]
cols = ['ColOne', 'ColTwo', 'ColThree', 'ColFour', 'ColFive']
rows = ['RowOne', 'RowTwo', 'RowThree']
fig = plt.figure()
ax = fig.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)

table = ax.table(cellText=data, colLabels=cols, rowLabels=rows, loc='upper center',cellLoc='center')
table.scale(1, 20)
```

Coloring table data can be done using below parameters:
* colColours - to color column header cells
* rowColours - to color row header cells
* cellColours - to color individual cells

```python
%matplotlib inline
import matplotlib.pyplot as plt

data=[
    [1, 2, 3, 4, 5],
    [2, 4, 6, 8, 10],
    [1, 3, 5, 7, 9]
]
cols = ['ColOne', 'ColTwo', 'ColThree', 'ColFour', 'ColFive']
rows = ['RowOne', 'RowTwo', 'RowThree']
colsc = ['gray', 'gray', 'gray', 'gray', 'gray']
rowsc = ['cyan', 'cyan', 'cyan']

colors = [["red","w","w","w","w"],[ "w","w","green","w","w"], [ "w","w","w","w","blue"]]

fig = plt.figure()
ax = fig.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)

table = ax.table(cellText=data, colLabels=cols, rowLabels=rows, loc='upper center',cellLoc='center', colColours=colsc, rowColours=rowsc, cellColours=colors)
table.scale(1, 20)
```

<blockquote>We can generate an image of our table using savefig() method.</blockquote>

```python
%matplotlib inline
import matplotlib.pyplot as plt

data=[
    [1, 2, 3, 4, 5],
    [2, 4, 6, 8, 10],
    [1, 3, 5, 7, 9]
]
cols = ['ColOne', 'ColTwo', 'ColThree', 'ColFour', 'ColFive']
rows = ['RowOne', 'RowTwo', 'RowThree']
colsc = ['gray', 'gray', 'gray', 'gray', 'gray']
rowsc = ['cyan', 'cyan', 'cyan']

colors = [["red","w","w","w","w"],[ "w","w","green","w","w"], [ "w","w","w","w","blue"]]

fig = plt.figure()
ax = fig.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)

table = ax.table(cellText=data, colLabels=cols, rowLabels=rows, loc='upper center',cellLoc='center', colColours=colsc, rowColours=rowsc, cellColours=colors)
table.scale(1, 20)

plt.savefig("pytable.png", format='png')
```
