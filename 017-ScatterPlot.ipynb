<b>Note: To plot bar chart, use plt.scatter() method</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

plt.scatter(x, y)
plt.show()
```

<b>Note: np.random.RandomState is from the numpy library to generate random numbers</b>

randint() methods used in below code accept three parameters.
* first parameter is the start number (low) to generate a random integer
* second parameter is the end nuber (high to generate a random integer
* third parameter is the total number of integer to be generated

```python
import numpy as np
rng = np.random.RandomState(0)

x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

plt.scatter(x,y)
plt.show()
```

<b>Note: Label can be used as same as other types of plot we have learned so far.</b>

```python
import numpy as np
rng = np.random.RandomState(0)

x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Students scores - Maths vs Physics')

plt.scatter(x,y)
plt.show()
```

Customization of plot style is possible with additional parameters
* s - sizes (can be single value or array like for size of data point)
* color - used to define the color of data point
* marker - used to define the style of data point

<b>Note: more style for marker can be found <a href="https://matplotlib.org/3.1.1/api/markers_api.html#module-matplotlib.markers">here</a> </b>

```python
import numpy as np
rng = np.random.RandomState(0)

x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

plt.scatter(x,y, s=100, color="green", marker="v")
plt.show()
```

<b>Note: Below example shows how to set the size of each data point</b><br/>
<b>Important: alpha is a blending value, between 0 (transparent) and 1 (opaque)</b>

```python
import numpy as np
rng = np.random.RandomState(0)
x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)

sizes = 10 * np.add(x, y)

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

<b>Below example shows how can we set different color or each data point (to make it more visible)</b>

```python
import numpy as np
rng = np.random.RandomState(0)
x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)
colors = rng.randint(1, 100, 50)
sizes = 10 * np.add(x, y)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)

plt.show()
```

<b>Note: Use plt.colorbar() method to add a color bar beside the plot</b>

```python
import numpy as np
rng = np.random.RandomState(0)
x = rng.randint(1, 100, 50)
y = rng.randint(1, 100, 50)
colors = rng.randint(1, 100, 50)
sizes = 10 * np.add(x, y)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='plasma')
plt.colorbar()
plt.show()
```

<b>Note: More information about colormap can be found <a href="https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html">here</a></b>
