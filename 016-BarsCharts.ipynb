<b>Note: To plot bar chart, use plt.bar() method</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = ['Samsung', 'Nokia', 'Apple', 'Oppo', 'Huawei', 'Sony']
y = [22, 12, 14, 6, 13, 18]
plt.bar(x, y)
plt.show()
```

To add more meaning to our bar chart, below methos are handy:
* plt.ylabel() - to add label to Y-Axis
* plt.xlabel() - to add label to X-Axis
* plt.title() - to add title of the bar chart

```python
plt.ylabel('Usage')
plt.xlabel('Mobile Phone')
plt.title('Usage of Mobile Phone Brand in a Classroom')
plt.bar(x, y)
plt.show()
```

<b>Tips: To plot horizontal bar chart, use plt.barh() method</b>

```python
plt.barh(x,y)
plt.show()
```

Note: subplots method provides a way to plot multiple plots on a single figure. Given the number of rows and columns , it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax

```python
import numpy as np

labels = ['Samsung', 'Nokia', 'Apple', 'Oppo', 'Huawei', 'Sony']
man = [10, 5, 4, 2, 10, 10]
woman= [12, 7, 10, 4, 3, 8]

fig, ax = plt.subplots()
x = np.arange(6)
width = 0.4

ax.bar(x - width/2, man, width, label='Men')
ax.bar(x + width/2, woman, width, label='Women')
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(labels)

plt.show()
```

<b>Note: In ax.bar() method, first parameter is the x-coordinates of the bar.</b>

To plot horizontal grouped bar chart, there are few things to take note:
* instead of calculating x-coordinate, we need to calculate on y
* instead of using set_xticks(), we need to use set_yticks()
* instead of using set_xticklabels(), we need to use set_yticklabels() 

```python
import numpy as np

labels = ['Samsung', 'Nokia', 'Apple', 'Oppo', 'Huawei', 'Sony']
man = [10, 5, 4, 2, 10, 10]
woman= [12, 7, 10, 4, 3, 8]

fig, ax = plt.subplots()
y = np.arange(6)
width = 0.4

ax.barh(y - width/2, man, width, label='Men')
ax.barh(y + width/2, woman, width, label='Women')
ax.legend()
ax.set_yticks(y)
ax.set_yticklabels(labels)

plt.show()
```

<b>Note: We can customize color of each bar by specifying addtional color parameter</b>

```python
plt.bar(labels, man, color="#00FF00")  
plt.bar(labels, woman, bottom=man, color="#FF0000")

plt.xlabel('Year')  
plt.ylabel('Number of futurestud.io Tutorials')
plt.title('Usage of Mobile Phone Brand in a Classroom')

plt.show()  
```

<b>Tips: To plot horizontal stacked bar chart, use plt.barh() method</b>

```python
plt.barh(labels, man, color="#00FF00")  
plt.barh(labels, woman, left=man, color="#FF0000")

plt.xlabel('Usage')  
plt.ylabel('Mobile Phone')
plt.title('Usage of Mobile Phone Brand in a Classroom')

plt.show()  
```
