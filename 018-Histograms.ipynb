Histogram can be plotted using plt.hist() methods with additional parameters
* first param is dataset, use list for multiple datasets
* second param is the number of bins to be used in the histogram (optional)

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
num_bins = 5
plt.hist(x, num_bins)
plt.show()
```

<b>Note: Below histogram used total of 7 bins</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
num_bins = 7
plt.hist(x, num_bins)
plt.show()
```

<b>Note: alpha is a blending value, between 0 (transparent) and 1 (opaque)</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
num_bins = 7
plt.hist(x, num_bins, color='red', alpha=0.5)
plt.show()
```

Aother useful parameter for styling histogram is "histtype"
* bar
* barstacked
* step
* stepfilled

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
num_bins = 7
plt.hist(x, num_bins, color='green', alpha=0.5, histtype='step')
plt.show()
```

<b>Note: Below histogram change orientation from vertical to horizontal</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
num_bins = 7
plt.hist(x, num_bins, color='green', alpha=0.5, histtype='bar', orientation ='horizontal')
plt.show()
```

It is possible to plot multiple datasets in a histogram, below are things to note:
* datasets need to given as list format
* labels optional but recommend to include to separate two datasets

```python
%matplotlib inline
import matplotlib.pyplot as plt

x1 = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
x2 = [20,20,30,30,45,45,50,50,65,60,65,75,75,80,80,90,90,35,85]

plt.hist([x1, x2],label=['x1', 'x2'])
plt.legend(loc='upper right')
plt.show()
```

It is possible to show one histogram over another.
* need to plot two histogram, i.e. calling plt.hist() two times with respective dataset
* use alpha transparency value to make mroe visible

<b>Note: if we plot two histogram, then we can set bin size for each separately</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x1 = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
x2 = [20,20,30,30,45,45,50,50,65,60,65,75,75,80,80,90,90,35,85]

plt.hist(x1, 5, label='x1', alpha=0.5, color='blue')
plt.hist(x2, 10, label='x2', alpha=0.5, color='red')
plt.legend(loc='upper right')
plt.show()
```

<b>Important: Labelling histogram is same as other plots</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x1 = [40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
x2 = [20,20,30,30,45,45,50,50,65,60,65,75,75,80,80,90,90,35,85]
num_bins = 7
plt.hist(x1, label='Maths', alpha=0.5, color='blue')
plt.hist(x2, label='Physics', alpha=0.5, color='red')
plt.xlabel('Marks')
plt.ylabel('Total Number')
plt.title('Students scores - Maths vs Physics')
plt.legend(loc='upper right')
plt.show()
```
