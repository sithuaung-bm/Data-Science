<b>Note: To plot line chart, use plt.plot() method</b>

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 3, 5]
y = [1, 3, 5]
plt.plot(x, y)
plt.show()
```

<b>Note: Use xlabel() and ylabel() to add title to each axis.</b>

```python
plt.xlabel('X-axis title')
plt.ylabel('Y-axis text')
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, linestyle='dotted', linewidth=1.5, color='red')
plt.show()
```

#### Line style can customize, some widely use line styles are:
* dashed
* dotted
* dashdot

```python
plt.plot(x, y, linestyle='dashed', linewidth=5, color='red')
plt.show()
```

```python
plt.plot(x, y, linestyle='dashdot', linewidth=2, color='red')
plt.show()
```

```python
import numpy as np

x = np.linspace(1, 10, 50)
x
```

```python
y = 2**x
print(y)
```

```python
plt.plot(x, y, label='Creating exponential line chart')
plt.legend()
plt.show()
```

<b>Important: Adding label parameter alone will not add legend in the chart, legend() must be called</b>

<b>Note: To plot more than one line in the chart, just call the plt.plot() method with different x & y value</b>

```python
x1 = [1, 3, 5, 7, 9]
y1 = [1, 3, 5, 7, 9]
x2 = [10, 8, 6, 4, 2]
y2 = [2, 4, 6, 8, 10]

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

<b>Note: To plot more than one line, we have to call plt.plot() mutliple time with respective x, y data</b>

```python
x1 = [1, 3, 5, 7, 9]
y1 = [1, 3, 5, 7, 9]
x2 = [10, 8, 6, 4, 2]
y2 = [2, 4, 6, 8, 10]
x3 = [1, 2, 3, 8]
y3 = [4, 5, 6, 7]

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```

<b>Note: To add data point as label, we can use plt.text() method with third parameters as string to show as data label</b>

```python
x=[1, 3, 5, 7, 9]
y=[2, 4, 6, 8, 10]

plt.plot(x,y)
for a,b in zip(x, y): 
    plt.text(a, b, str(a) + ', ' + str(b))
plt.show()
```

<b>Above example show how to add data point as lable in line chart</b>
